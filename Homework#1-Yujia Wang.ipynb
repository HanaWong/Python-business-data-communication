```python
# Homework 1

### Copy and paste all of this code into your program. 
### Please do not change any of this code. If you accidentally change something,
### come back to this notebook and copy a clean version. 

from random import randint
animals = ['tiger', 'lion', 'elephant', 'giraffe', 'leopard', 'panda', 'cheetah', 'koala', 'gorilla', 'donkey']

target_word = animals[randint(0, 9)]

### START YOUR OWN CODE BELOW THIS LINE ###

incorrect_list = []

target_word_list = []
for letter in target_word:
    target_word_list.append(letter)
    
word_length = len(target_word)

user_guessed = ['-' for _ in range(word_length)]
print("The Word Is:", user_guessed)


def judge(x, count):
    if len(x) ==1:
        found = False
        for index in range(len(target_word_list)):
            if target_word_list[index] == x:
                found = True
                user_guessed[index] = x

        if not found:
            incorrect_list.append(x)
            print("The Word is:",user_guessed)
            print("Incorrect, you have tried:",incorrect_list)
        else:
            print("Your guess is correct:",user_guessed)
            print("You have tried these letters but they are not correct:",incorrect_list)
        count = count - 1
        print("Remaining times:", count)
    
    elif x == target_word:
        print("You win! The Word Is:", target_word)
        count = -1
    else:   
        print("Incorrect, you lose the game, the word is:",target_word_list)
        count = 0
        print ("Remaining times: ",count)
        
    return count
        
time = 10
while time > 0:
    player = input("Please Guess A Letter or The Word:") 
    time = judge(player, time)
    print()
   
    if time == 0:
        print("Incorrect, you have no remaining time, you lose the game.")
        break
    
    elif time == -1:
        break

```
