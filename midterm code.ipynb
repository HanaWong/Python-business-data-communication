# Midterm- Making appointment

```python
dayMap = {'1': 'Monday', '2': 'Tuesday', '3': 'Wednesday', '4': 'Thursday'}
appointed = {}

def printAppointment(appointed, dayMap):
    if len(appointed.keys()) == 0:
        print('Monday\n\nTuesday\n\nWednesday\n\nThursday\n\n')
        return
    space = len('Wednesday') 
  
    monAppoints = []
    tueAppoints = []
    wedAppoints = []
    thurAppoints = []

    for key, name in appointed.items():
        day, time = key[0], key[1]
        if day == '1':
            monAppoints.append((time, name))
        elif day == '2':
            tueAppoints.append((time, name))
        elif day == '3':
            wedAppoints.append((time, name))
        elif day == '4':
            thurAppoints.append((time, name))
  
    for j, appoints in enumerate([monAppoints, tueAppoints, wedAppoints, thurAppoints]):
        day = dayMap[str(j+1)]
        if not appoints:
            print(day)
            continue
        appoints.sort(key=lambda x:x[0])
        for i in range(len(appoints)):
            time, name = appoints[i][0],  appoints[i][1]
            if i == 0:
                print('{}{}\t{}\t{}'.format(day, ' '*(space-len(day)), time, name))
            else:
                print('{}\t{}\t{}'.format(' '*space,time, name))

    print('\n')
    return

while True:
    choice = input('Would you like to (s)chedule, view (a)dmin or (q)uits? ')
    if choice == 'q':
        print('Goodbye!')
        break
  
    elif choice == 'a':
        printAppointment(appointed, dayMap)
  
    elif choice == 's':
        while True:
            day = input('Here are the days you can choose from:\n1 . Monday\n2 . Tuesday\n3 . Wednesday\n4 . Thursday\nPlease enter the number for the day you\'d like to choose for your meeting (1-4): ')
            time = input('Here are the meeting times you can choose from:\n1 . 2:30\n2 . 3:00\n3 . 3:30\n4 . 4:00\n5 . 4:30\nPlease enter the time you wish to schedule for (ex: 3:30): ')
            if (day, time) in appointed.keys():
                print('This time slot has been taken. Please choose another one\n')
                continue
            else:
                name = input('Please enter your name(s): ')
                appointed[(day, time)] = name
                print('Hi {}, your appointment at {} {} has been confirmed\n'.format(name, dayMap[day], time))
                break
      
```
